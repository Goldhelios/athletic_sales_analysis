### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrame
df_2021 = pd.read_csv('athletic_sales_2021.csv')
df_2020 = pd.read_csv('athletic_sales_2020.csv')
```

#### Combine the sales data by rows.

```python
# Combine df_2020 and df_2021 DataFrames on the rows and reset the index.
combined_df = pd.concat([df_2020, df_2021], ignore_index=True)
combined_df
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
import pandas as pd

# Create pivot table for number of products sold by region
pivot_table = pd.pivot_table(combined_df, index=['region','state','city'], values='product', aggfunc='count')
pivot_table = pd.DataFrame(pivot_table)
pivot_table
```

```python
# Rename the product column to "Total_Products_Sold" in the dataframe pivot_table
pivot_table.rename(columns={'product': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
pivot_table.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
# Rename the "units_sold" column to "Total_Products_Sold"

grouped_qty = combined_df.groupby(['retailer', 'state', 'city'])\
        .agg({'units_sold': 'sum'})\
        .reset_index()\
        .rename(columns={'units_sold': 'Total_Products_Sold'})



# Show the top 5 results.
grouped_qty.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_df = combined_df[combined_df['product'] == "Women's Athletic Footwear"]
womens_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold in womens_df for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
grouped_sales_women = womens_df.groupby(['retailer', 'region', 'state', 'city'])\
        .agg({'units_sold': 'sum'})\
        .rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
grouped_sales_women.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_sales_women = womens_df.pivot_table(index=['retailer', 'region', 'state','city'], 
                                            values='units_sold', 
                                            aggfunc='sum')\
                        .rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})\

# Show the top 5 results.
pivot_sales_women.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

pivot_date_women = womens_df.pivot_table(index=['invoice_date'], 
                                            values='total_sales', 
                                            aggfunc='sum')\
        .rename(columns={'total_sales':'Total Sales'})
                        
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the table.
pivot_date_women.head()
```

```python
# Resample the pivot_date_women df into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in descending order on "Total Sales".

women_daysample =pivot_date_women.resample('D').sum()
women_daysample.sort_values(by="Total Sales", ascending=False)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Sort the resampled pivot table in descending order on "Total Sales".

women_wksample = pivot_date_women.resample('W').sum()
women_wksample.sort_values(by="Total Sales", ascending=False)
```
